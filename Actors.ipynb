```python
import cv2
import sys
import time

cascPath = "haarcascade_frontalface_default.xml"
faceCascade = cv2.CascadeClassifier(cascPath)

imagex = 100
imagey = 100
differentpeople = 2
```

```python
import glob, os

opening = False

for i in range(0,1):
    path = 'C:\\Users\\Andreas\\Desktop\\Programme\\1----Python programms\\Actors_Keras\\Actors\\*'
    for infile in glob.glob(path):
        for infile2 in glob.glob(infile + "\\*"):
            WindowOpen = False
            try:
                img = cv2.imread(infile2)
                gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                faces = faceCascade.detectMultiScale(gray, 1.3, 5)
            except:
                pass
            
            count = 0
            for (x,y,w,h) in faces:
                count += 1
                if(opening):
                    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
                    roi_gray = gray[y:y+h, x:x+w]
                    roi_color = img[y:y+h, x:x+w]
                    cv2.imshow('img',img) #[y:y+h, x:x+w]
                WindowOpen = True
                if(count > 1):
                    WindowOpen = False
                
            if(WindowOpen):
                WindowOpen = False
                if(opening):
                    cv2.waitKey(0)
                    cv2.destroyAllWindows()
            else:
                print('Removing: ' + infile2)
                os.remove(infile2)

print('Finished')
```

```python
def GetFace(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = faceCascade.detectMultiScale(gray, 1.3, 5)
        
    for (x,y,w,h) in faces:
        img = cv2.resize(img[y:y+w, x:x+w], dsize=(imagex, imagey), interpolation=cv2.INTER_CUBIC)
        #cv2.imshow('img',img)
        #cv2.waitKey(0)
        #cv2.destroyAllWindows()
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        return img
```

```python
import numpy as np
import cv2
import glob, os
import random

#randseed = random.randint(0, 100000)

res = GetFace(cv2.imread('C:\\Users\\Andreas\\Desktop\\Programme\\1----Python programms\\Actors_Keras\\Actors\\aaron eckhart\\0.jpg')).reshape(1,imagex,imagey,1)
#res = np.resize(res, (1, 1000))

y = np.zeros(shape=(1,differentpeople))
y = y.reshape(1, differentpeople)
y[0][0] = 1

path = 'C:\\Users\\Andreas\\Desktop\\Programme\\1----Python programms\\Actors_Keras\\Actors\\*'
count = 0
for infile in glob.glob(path):  
    for infile2 in glob.glob(infile + "\\*"):
        try:
            res = np.append(res, GetFace(cv2.imread(infile2)).reshape(1,imagex,imagey,1), axis=0)
            yappend = np.zeros(shape=(1,differentpeople))
            yappend[0][count] = 1
            y = np.append(y, yappend, axis=0)
        except:
            pass

    print(infile)
    count += 1

#np.random.seed(randseed)
#np.random.shuffle(res)
#np.random.shuffle(y)
    
print(res.shape)
print(y.shape)
```

```python
import tensorflow.keras as ks
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten
from tensorflow.keras.layers import Dropout

model = ks.Sequential([
    #Dense(units=imagex*imagey, activation='relu', input_dim=imagex*imagey),
    Conv2D(64, (3, 3), input_shape=(imagex, imagey, 1), activation='relu'),
    Conv2D(64, (3, 3), padding="same", activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(128, (3, 3), input_shape=(imagex, imagey, 1), activation='relu'),
    Conv2D(128, (3, 3), padding="same", activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(256, (3, 3), input_shape=(imagex, imagey, 1), activation='relu'),
    Conv2D(256, (3, 3), padding="same", activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(units=512, activation='relu'),
    Dense(units=256, activation='relu'),
    Dense(units=128, activation='relu'),
    Dense(units=differentpeople, activation='softmax')
])

adam2 = ks.optimizers.Adam(lr=0.00005)
sgd = ks.optimizers.SGD(lr=0.0001, decay=1e-6, momentum=0.9, nesterov=True)

model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])

model.summary()
```

```python
model.fit(res, y, epochs=10, batch_size=50)
```
